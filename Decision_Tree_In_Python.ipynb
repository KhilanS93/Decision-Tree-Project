```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

```python
from sklearn.model_selection import train_test_split #import to train and  test out data
x=dataset.copy() #x gets whole data set with target answers so that it can learn itself
y=data['target'] #y gets the target values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33) # seprate training data and keep 33 % of data as test data
```

```python
from sklearn.tree import DecisionTreeClassifier #import the model itself
clf= DecisionTreeClassifier(max_depth=4) #create classfifer and input the parameter max depth for deep the tree can be
clf=clf.fit(x_train,y_train)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(x_test)
predictions
```

```python
clf.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
print(accuracy_score(y_test,predictions))
print(confusion_matrix(y_test,predictions,labels=[0,1]))
print(precision_score(y_test,predictions))
print(recall_score(y_test,predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions,target_names=['malignant','benign']))
```
